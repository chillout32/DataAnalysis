### 5. I Python ska du nu:
### a) Skapa en klass som heter BankAccount. Klassen skall ha attributet "account_holder"
### som visar kontoinnehavarens namn samt attributet "balance" som visar
### kontoinnehavarens balans. Klassen skall ha metoden "deposit()" för att kunna sätta in
### pengar på kontot samt metoden "withdraw()" för att kunna ta ut pengar från kontot. Om
### bankinnehavaren försöker ta ut mer pengar än vad som finns på kontot skall
### meddelandet "Too low balance" printas ut.
### b) Skapa en instans av klassen och testa så klassen funkar så som du förväntar dig. Du
### kan till exempel prova printa ut attributen, sätta in pengar och ta ut pengar.

```python
class BankAccount:
    def __init__(self, account_holder):
        self.account_holder = account_holder
        self.balance = 0.0

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"{amount} deposited. New balance: {self.balance}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Too low balance")
        elif amount <= 0:
            print("Withdrawal amount must be positive.")
        else:
            self.balance -= amount
            print(f"{amount} withdrawn. New balance: {self.balance}")
```

```python
account = BankAccount("Anna Andersson")
account.deposit(1000)
account.withdraw(300)
account.withdraw(800)  # Försök ta ut mer än vad som finns kvar
```

### 6. På intervjuer brukar ibland kodtest genomföras. Här är ett exempel på en fråga som hade
### kunnat dyka upp på ett kodtest.
### Skapa en Python funktion som beräknar antalet vokaler i en sträng. Vokalerna är:
### A, E, I, O, U, Y, Å, Ä och Ö.

```python
def count_vowels(text):
    vowels = "AEIOUYÅÄÖaeiouyåäö"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

# Exempelanvändning:
test_str = "Hej hur många vokaler finns i denna text?"
print("Antal vokaler:", count_vowels(test_str))
```

### 7. På intervjuer brukar ibland kodtest genomföras. Här är ett exempel på en fråga som hade
### kunnat dyka upp på ett kodtest.
### Skapa en funktion som tar två listor som argument och returnerar de gemensamma
### elementen i listorna i en ny lista.
### Exempelvis:
### common_elements([4, ’apple’, 10, ’hi’, 3], ([23 ’apple’, 5, 9, 3]) ska returnera svaret
### [’apple’, 3].

```python
def common_elements(list1, list2):
    return [element for element in list1 if element in list2]

resultat = common_elements([4, 'apple', 10, 'hi', 3], [23, 'apple', 5, 9, 3])
print("Gemensamma element:", resultat)
```

# Del 2 – Dataanalys i Python

## We are analyzing diamonds with the intentions of investing, we will be analyzing different combinations and mainly looking at what affects price. Trying to find the best investment.

## We will be using the following format:
### From this chart we can see...
### The following implications:
### This means that...

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("diamonds.csv", sep=",")
```

```python
data['volume'] = data['x'] * data['y'] * data['z']

numeric_data = data.select_dtypes(include=[np.number])
corr_matrix = numeric_data.corr()

plt.figure(figsize=(10, 8))
plt.imshow(corr_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation=45, ha='right')
plt.yticks(range(len(corr_matrix)), corr_matrix.columns)
plt.title("Correlation Matrix for the diamonds")
plt.tight_layout()
plt.show()
```

### From this chart, we can see a general heat map of different combinations and their correlation. This is amazing just to see a general overview of diamonds.

### The following implications:
### * Volume/carat affects price greatly

### This means that using the help of this graph you can just get some general knowledge about diamonds and what affects the price.

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['carat'], data['price'])
plt.xlabel("Carat")
plt.ylabel("Price (USD)")
plt.title("Price vs Carat")
plt.grid(True)
plt.tight_layout()
plt.show()
```

### From this chart, we can see a scatter plot that analyzes price vs carat, we can clearly see that carat greatly affects price.
### The following implications:
### * Higher carat diamonds are more expensive
### * Lower carat diamonds are less expensive on average

### This means that there are several factors that affects price, not just carat, but generally higher carat = higher price

```python
avg_price_by_cut = data.groupby('cut')['price'].mean()

plt.figure(figsize=(8, 6))
plt.bar(avg_price_by_cut.index, avg_price_by_cut.values, color='lightgreen')
plt.xlabel("Cut")
plt.ylabel("Average Price (USD)")
plt.title("Average Diamond Price by Cut")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### From this chart, we can see the average price of diamonds based on their cut,we can see that ideal cut diamonds on average are the cheapest, and premium cut diamonds are the most expensive

### The following implications:
### * Premium diamonds are the most expensive
### * Ideal cut diamonds are the least expensive

### This means that when investing this gives a good overview which diamonds to invest in based on cut.

```python
# Converts to string
data['color'] = data['color'].astype(str)
data['clarity'] = data['clarity'].astype(str)

colors = sorted(data['color'].unique())
clarities = sorted(data['clarity'].unique())

# Average price for each combination
grouped = data.groupby(['color', 'clarity'])['price'].mean().unstack()

# Plot grouped bar chart
x = np.arange(len(clarities))
bar_width = 0.1

plt.figure(figsize=(12, 6))

for i, color in enumerate(colors):
    prices = grouped.loc[color]
    plt.bar(x + i * bar_width, prices.values, width=bar_width, label=f'Color {color}')

plt.xlabel("Clarity")
plt.ylabel("Average Price (USD)")
plt.title("Average Diamond Price by Clarity and Color")
plt.xticks(x + bar_width * (len(colors) / 2), clarities, rotation=45)
plt.legend(title="Color")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

### From this chart, we can see prices of different combinations of Clarity and Color, we can clearly see that Color D with the clarity level IF is the most expensive.

### The following implications:
### * High clarity and color = highest value
### * Color impacts price more at higher clarity levels
### * At lower clarity, color matters less

### This means that for best investment value, choose diamonds with top clarity and color. For budget options, prioritize color over clarity.

```python
data = data.copy()
data['clarity'] = data['clarity'].replace(['', ' ', 'NaN', 'nan'], np.nan) # Converts empty data to NaN so we can hide it
plt.figure(figsize=(10, 6))
data.boxplot(column='carat', by='clarity', grid=False)
plt.title("Carat vs Clarity")
plt.suptitle("")
plt.xlabel("Clarity")
plt.ylabel("Carat")
plt.tight_layout()
plt.show()
```

### From this chart, we can see if clarity and carat has some sort of correlation, they don't really have a clear correlation but still provides good insight based on the data.

### The following implications:
### * l1 diamonds have the highest spread in carat

### This means that carat most likely is not affected by clarity and that you should thoroughly look at these factors when choosing your investments.

```python
# drop rows with missing or zero dimensions or price
data = data[(data['x'] > 0) & (data['y'] > 0) & (data['z'] > 0) & (data['price'] > 0)]

# Scatter plots for each dimension vs price
fig, axs = plt.subplots(1, 3, figsize=(18, 5))
axs[0].scatter(data['x'], data['price'], alpha=0.5)
axs[0].set_xlabel('Length (x mm)')
axs[0].set_ylabel('Price (USD)')
axs[0].set_title('Price vs Length')

axs[1].scatter(data['y'], data['price'], alpha=0.5, color='orange')
axs[1].set_xlabel('Width (y mm)')
axs[1].set_ylabel('Price (USD)')
axs[1].set_title('Price vs Width')

axs[2].scatter(data['z'], data['price'], alpha=0.5, color='green')
axs[2].set_xlabel('Depth (z mm)')
axs[2].set_ylabel('Price (USD)')
axs[2].set_title('Price vs Depth')

plt.tight_layout()
plt.show()

# Correlation between price and dimensions
corr = data[['price', 'x', 'y', 'z']].corr()
print("Correlation matrix:\n", corr)
```

### From this chart, we can see how length, width and depth affects price, where length has the biggest price spread.

### The following implications:
### * Length is the one that affects price the most.
### * Width and depth has roughly the same impact on price, and the impact is not huge.

### This means that when looking for investments, wide and deep diamonds are the most consistent whereas length is the least consistent.

```python
data = data[(data['price'] > 0) & (data['table'] > 0) & (data['depth'] > 0)]

# 2 plots
plt.figure(figsize=(14, 5))

# Table vs Price
plt.subplot(1, 2, 1)
plt.scatter(data['table'], data['price'], alpha=0.5, color='purple')
plt.xlabel("Table (%)")
plt.ylabel("Price (USD)")
plt.title("Price vs Table Size")
plt.grid(True, linestyle='--', alpha=0.6)

# Depth vs Price
plt.subplot(1, 2, 2)
plt.scatter(data['depth'], data['price'], alpha=0.5, color='teal')
plt.xlabel("Depth (%)")
plt.ylabel("Price (USD)")
plt.title("Price vs Depth")
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

### From this chart, we can see charts of price vs the table (%) and the depth (%) and analyze if they impact price.
#### Depth is calculated by a formula using the x, y and z parameters of the diamond, the table's size is expressed as a percentage of the diamond's average girdle diameter.
### The following implications:
### * Moderate depth (%) tends to correlate better with higher prices extremely shallow or deep stones may be less valuable.
### * Table size (%) also shows a price pattern, with mid-range values often attracting higher prices.
### This means that when looking for investments, wide and deep diamonds are the most consistent whereas length is the least consistent.
